```python
import glob,os
import cv2
import numpy as np
import matplotlib.pyplot as plot
#%matplotlib inline
from IPython.display import Image
#from keras.models import Model
#from keras.layers import Flatten,Dense
#from keras.applications.vgg16 import VGG16 #Import all the necessary modules
```

```python


#with pneumonia train image
img_trainp=[]
img_trainn=[]
for img in glob.glob("./chest-xray/train/PNEUMONIA/*"):
    jpg = cv2.imread(img)
    image_bw = cv2.cvtColor(jpg, cv2.COLOR_BGR2GRAY)

    clahe = cv2.createCLAHE(clipLimit=2.5)
    final_img = clahe.apply(image_bw) + 12
    img_trainp.append(final_img)
    #print(img_trainp)

for img1 in glob.glob("./chest-xray/train/NORMAL/*"):
    jpg1 = cv2.imread(img1)
    image_bw1 = cv2.cvtColor(jpg1, cv2.COLOR_BGR2GRAY)

    clahe1 = cv2.createCLAHE(clipLimit=2.5)
    final_img1 = clahe1.apply(image_bw1) + 12
    img_trainn.append(final_img1)
    #print(img_trainp)
    
#Test image Processing
img_testp=[]
img_testn=[]
for img in glob.glob("./chest-xray/test/PNEUMONIA/*"):
    jpg = cv2.imread(img)
    image_bw = cv2.cvtColor(jpg, cv2.COLOR_BGR2GRAY)

    clahe = cv2.createCLAHE(clipLimit=2.5)
    final_img = clahe.apply(image_bw) + 12
    img_testp.append(final_img)
    #print(img_trainp)

for img1 in glob.glob("./chest-xray/test/NORMAL/*"):
    jpg1 = cv2.imread(img1)
    image_bw1 = cv2.cvtColor(jpg1, cv2.COLOR_BGR2GRAY)

    clahe1 = cv2.createCLAHE(clipLimit=2.5)
    final_img1 = clahe1.apply(image_bw1) + 12
    img_testn.append(final_img1)
    #print(img_trainp)


#Validation Image Processing
img_valp=[]
img_valn=[]
for img in glob.glob("./chest-xray/val/PNEUMONIA/*"):
    jpg = cv2.imread(img)
    image_bw = cv2.cvtColor(jpg, cv2.COLOR_BGR2GRAY)

    clahe = cv2.createCLAHE(clipLimit=2.5)
    final_img = clahe.apply(image_bw) + 12
    img_valp.append(final_img)
    #print(img_trainp)

for img1 in glob.glob("./chest-xray/val/NORMAL/*"):
    jpg1 = cv2.imread(img1)
    image_bw1 = cv2.cvtColor(jpg1, cv2.COLOR_BGR2GRAY)

    clahe1 = cv2.createCLAHE(clipLimit=2.5)
    final_img1 = clahe1.apply(image_bw1) + 12
    img_valn.append(final_img1)


```

```python
import glob,os
import cv2
img_trainpc=[]
img_trainnc=[]
i=1
for img in glob.glob("chest_xray/update_train/PNEUMONIA/*"):
    jpg = cv2.imread(img)
    gray = cv2.cvtColor(jpg, cv2.COLOR_BGR2GRAY)
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(contours) < 2:
        print("Error: Image does not have at least two contours")
        print(i)
        i=i+1
    else:
        contours = sorted(contours, key=cv2.contourArea, reverse=True)
        largest_contour = contours[0]
        second_largest_contour = contours[1]
        x1, y1, w1, h1 = cv2.boundingRect(largest_contour)
        x2, y2, w2, h2 = cv2.boundingRect(second_largest_contour)
        cropped_image = jpg[min(y1, y2):max(y1 + h1, y2 + h2), min(x1, x2):max(x1 + w1, x2 + w2)]
        img_trainpc.append(cropped_image)
        i=i+1
print(len(img_trainpc))
i=1
for img in glob.glob("chest_xray/update_train/NORMAL/*"):
    jpg = cv2.imread(img)
    gray = cv2.cvtColor(jpg, cv2.COLOR_BGR2GRAY)
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(contours) < 2:
        print("Error: Image does not have at least two contours")
        print(i)
    else:
        contours = sorted(contours, key=cv2.contourArea, reverse=True)
        largest_contour = contours[0]
        second_largest_contour = contours[1]
        x1, y1, w1, h1 = cv2.boundingRect(largest_contour)
        x2, y2, w2, h2 = cv2.boundingRect(second_largest_contour)
        cropped_image = jpg[min(y1, y2):max(y1 + h1, y2 + h2), min(x1, x2):max(x1 + w1, x2 + w2)]
        img_trainnc.append(cropped_image)
        i=i+1
print(len(img_trainnc))
i=1
for img in img_trainnc:
    filename = "train_normal_crop_img_c"+str(i)+".jpeg"
    path = r"C:/Users\abakf\Desktop\Project\pneumonia-model\chest_xray\update_train_c\NORMAL"
    cv2.imwrite(os.path.join(path , filename), img)
    i=i+1
i=1
for img in img_trainpc:
    filename = "train_pneumonia_crop_img_c"+str(i)+".jpeg"
    path = r"C:/Users\abakf\Desktop\Project\pneumonia-model\chest_xray\update_train_c\PNEUMONIA"
    cv2.imwrite(os.path.join(path , filename), img)
    i=i+1
```
